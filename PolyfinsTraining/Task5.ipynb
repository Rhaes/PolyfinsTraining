```python
import pandas as pd
import numpy as np
import statsmodels.tsa.api as tsa
```

```python
# Read the daily-min-temperatures.csv dataset into a Pandas DataFrame
df_temps = pd.read_csv('daily-min-temperatures.csv', parse_dates=[0])
df_temps.rename(columns={'Temp': 'temperature'}, inplace=True)
```

```python
# Split the dataset into the training and testing sets
train_data = df_temps[0:int(0.8 * len(df_temps))]
test_data = df_temps[int(0.8 * len(df_temps)):]
```

```python
# Explore the daily-min-temperatures.csv dataset
print(df_temps.describe())
df_temps.plot()
```

```python
# Apply the AR model to the training set
ar_model = tsa.AR(train_data['temperature'])
ar_model_fit = ar_model.fit()

# Apply the MA model to the training set
ma_model = tsa.MA(train_data['temperature'])
ma_model_fit = ma_model.fit()

# Apply the ARMA model to the training set
arma_model = tsa.ARMA(train_data['temperature'], order=(1, 1))
arma_model_fit = arma_model.fit()

# Apply the ARIMA model to the training set
arima_model = tsa.ARIMA(train_data['temperature'], order=(1, 1, 1))
arima_model_fit = arima_model.fit()

# Evaluate the performance of the models on the testing set
ar_predictions = ar_model_fit.predict(start=len(train_data), end=len(df_temps))
ma_predictions = ma_model_fit.predict(start=len(train_data), end=len(df_temps))
arma_predictions = arma_model_fit.predict(start=len(train_data), end=len(df_temps))
arima_predictions = arima_model_fit.predict(start=len(train_data), end=len(df_temps))

# Calculate the mean squared error (MSE) for each model
ar_mse = np.mean((ar_predictions - test_data['temperature'])**2)
ma_mse = np.mean((ma_predictions - test_data['temperature'])**2)
arma_mse = np.mean((arma_predictions - test_data['temperature'])**2)
arima_mse = np.mean((arima_predictions - test_data['temperature'])**2)

# Print the MSE for each model
print('AR MSE:', ar_mse)
print('MA MSE:', ma_mse)
print('ARMA MSE:', arma_mse)
print('ARIMA MSE:', arima_mse)
```
